```python
dt=1
a=0.05
b=0.005
c=0.1
d=0.005


import numpy as np

import matplotlib.pyplot as plt
```

```python
def proies(n1,n2,e):
    return a*n1-b*n1*n2-e*n1
```

```python
def prédateurs(n1,n2,e):
    return -c*n2+d*n1*n2-e*n2
```

```python
def liste(n1,n2,tmin,tmax,e):

    lproies=[]
    lpred=[]
    lt=[]
    t=tmin
    for i in range(tmax):
        lproies.append(n1)
        n1=proies(n1,n2,e)*dt+n1
        lpred.append(n2)
        n2=prédateurs(n1,n2,e)*dt+n2
        lt.append(t)
        t=t+1
    return lt,lproies,lpred    
```

```python

def affiche(n1,n2,tmin,tmax,emin,emax):
    
    Tliste=[]
    Lliste=[]
    Yliste=[]
    
    for e in np.linspace(emin,emax,num=10):

        T,L,Y=liste(n1,n2,tmin,tmax,e)
        Tliste.append(T)
        Lliste.append(L)
        Yliste.append(Y)
        
    for j in range(len(Tliste)):

        plt.plot(Tliste[j],Lliste[j])

    plt.xlabel('Temps')

    plt.ylabel('nombre de proies')
    

    plt.show()
    
    
    for h in range(len(Tliste)):

        plt.plot(Tliste[h],Yliste[h])

    plt.xlabel('Temps')

    plt.ylabel('nombre de predateurs')
    

    plt.show()
    
    for g in range(len(Lliste)):
        
        plt.plot(Lliste[g],Yliste[g])

    plt.xlabel('nombre de proies')

    plt.ylabel('nombre de predateurs')
    

    plt.show()

```

```python
affiche(30,40,0,500,0.01,0.1)
```
