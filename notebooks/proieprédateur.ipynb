```python
dt=1
a=0.05
b=0.005
c=0.1
d=0.005


import numpy as np

import matplotlib.pyplot as plt
```

```python
def proies(n1,n2):
    return a*n1-b*n1*n2
```

```python
def prédateurs(n1,n2):
    return -c*n2+d*n1*n2
```

```python
def liste(n1,n2,tmin,tmax):

    lproies=[]
    lpred=[]
    lt=[]
    t=tmin
    for i in range(tmax):
        lproies.append(n1)
        n1=proies(n1,n2)*dt+n1
        lpred.append(n2)
        n2=prédateurs(n1,n2)*dt+n2
        lt.append(t)
        t=t+1
    return lt,lproies,lpred    
```

```python

def afficheproies(n1,n2,tmin,tmax
    plt.show()):

    T,L,Y=liste(n1,n2,tmin,tmax)

    plt.plot(T,L)

    plt.xlabel('Temps')

    plt.ylabel('nombre de proies')

    plt.show()

```

```python
afficheproies(30,40,0,500)
```

```python
def affichepred(n1,n2,tmin,tmax):

    T,L,Y=liste(n1,n2,tmin,tmax)

    plt.plot(T,Y)

    plt.xlabel('Temps')

    plt.ylabel('nombre de predateurs')

    plt.show()

```

```python
affichepred(30,40,0,500)
```
